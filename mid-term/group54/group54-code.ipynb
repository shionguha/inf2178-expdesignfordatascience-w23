```python
# https://medium.com/analytics-vidhya/testing-a-difference-in-population-proportions-in-python-89d57a06254
import scipy.stats as stats
import math

# Specify the number of occurences (x), the sample size (n), and the proportion claimed in the null-hypothesis (p)
x = female['Fraud'] # 134
n = national['Fraud'] # 474
p = female.sum()/national.sum() # 0.19329667291327457

# H0: Prob Fraud Female == 0.19
# Ha: Prob Fraud Female != 0.19

# Calculate the sample proportion
p_hat = x/n

# Calculate the test statistic
test_stat = (p_hat-p)/(math.sqrt((p*(1-p))/(n)))

# Output the p-value of the test statistic (two-tailed test)
p = 2*stats.norm.cdf(-abs(test_stat))

test_stat, p
```

## One-way ANOVA test for crimes per division

```python
df['date'] = df.Arrest_Year.apply(str) + '-' + df.Arrest_Month.replace({'Jan-Mar': 'Q1', 'Apr-June': 'Q2', 'July-Sept': 'Q3', 'Oct-Dec': 'Q4'})
df = df[['date'] + [c for c in df.columns if c != 'date']]
cdf = df[df.ArrestLocDiv != 'XX']
cdf.ArrestLocDiv.unique()
```

```python
cdf[cdf.ArrestLocDiv == '54'].groupby('date').ObjectId.nunique().tolist()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

divisions = sorted(cdf.ArrestLocDiv.unique())
for div in divisions:
  y = cdf[cdf.ArrestLocDiv == div].groupby('date').ObjectId.nunique().tolist()
  fig.add_trace(go.Box(y=y, name=div))

fig.update_layout(
    yaxis_title='Number Arrests Per Quarter',
    xaxis_title='Police Division'
)
fig.show()
```

```python
from scipy.stats import f_oneway

# prepare data
samples = []
divisions = sorted(cdf.ArrestLocDiv.unique())
for div in divisions:
  samples.append(cdf[cdf.ArrestLocDiv == div].groupby('date').ObjectId.nunique().tolist())

# One-way ANOVA
statistic, pvalue = f_oneway(*samples)

print(f'One-way ANOVA: s = {statistic}, p = {pvalue}')
```

```python
def get_significance(p):
    """Returns the significance of a p-values as a string of stars."""
    if p <= 0.001:
        return '***'
    elif p <= 0.01:
        return '**'
    elif p <= 0.05:
        return '*'
    elif p <= 0.1:
        return '.'
    else:
        return ''


def round_p_value(p):
    """Round a small p-value so that it is human-readable."""
    if p < 0.001:
        return '< 0.001'
    else:
        return f'{p:5.3}'


p_rounded = round_p_value(pvalue)
significance = get_significance(pvalue)
print(f'The p-value is {p_rounded} ({significance})')
```

```python
rows = []
for sample, division in zip(samples, divisions):
  for s in sample:
    rows.append([s, division])
  
tukey_df = pd.DataFrame(columns=['arrests', 'division'], data=rows)
tukey_df
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# perform Tukey's test
tukey = pairwise_tukeyhsd(endog=tukey_df['arrests'],
                          groups=tukey_df['division'],
                          alpha=0.05)

#display results
print(tukey)
```

## Two-way ANOVA test for interaction effect between division and race

```python
df['date'] = df.Arrest_Year.apply(str) + '-' + df.Arrest_Month.replace({'Jan-Mar': 'Q1', 'Apr-June': 'Q2', 'July-Sept': 'Q3', 'Oct-Dec': 'Q4'})
df = df[['date'] + [c for c in df.columns if c != 'date']]
cdf = df[(df.ArrestLocDiv != 'XX') & (df.Perceived_Race != 'Unknown or Legacy') & ~df.Perceived_Race.isnull()]
cdf.Perceived_Race.unique()
```

```python
grouped = cdf.groupby(['ArrestLocDiv', 'Perceived_Race', 'date'])\
             .ObjectId.nunique()\
             .reset_index()
grouped.columns = ['division', 'ethnicity', 'quarter', 'arrests_per_quarter']
grouped = grouped.drop('quarter', axis=1)
grouped
```

```python
import plotly.express as px

fig = px.box(grouped, x="division", y="arrests_per_quarter", color="ethnicity")
fig.update_traces(quartilemethod="exclusive")
fig.show()
```

```python
import plotly.express as px

fig = px.box(grouped, x="ethnicity", y="arrests_per_quarter", color="division")
fig.update_traces(quartilemethod="exclusive")
fig.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('arrests_per_quarter ~ C(ethnicity) + C(division) + C(ethnicity):C(division)', data=grouped).fit()
sm.stats.anova_lm(model, typ=2)
```

```python
from statsmodels.graphics.factorplots import interaction_plot

fig = interaction_plot(grouped.division, grouped.ethnicity, grouped.arrests_per_quarter, ms=10)
plt.legend(bbox_to_anchor=(1, 0.5))
```
