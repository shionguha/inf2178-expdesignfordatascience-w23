# Load the dataset 

```python
df = pd.read_csv('student-mat.csv')
```

```python
df.head()
```

# Run One way ANOCOVA 

```python
df['Mjob'].unique() #our x_1 which is a categorical variable 
```

```python
df['G1'].unique() #our x_2 is NOT a categorical variable and is another continuous variable 
```

```python
df['G3'].unique() #our dependent variable 
```

```python
#isolate columns of interest into a new df
df_1way = df[['Mjob', 'G1', 'G3']].copy()

df_1way.head()
```

```python
#inspect the data is stored correctly in the df
df_1way.info()
```

The null hypothesis for this example: Students from all mother job groups have different final grades on average after accounting for their first term grade.

```python
#what is the conditional proability of predicitng y = p(x_1|x_2)
ancova(data=df_1way, dv='G3', covar='G1', between='Mjob')
```

**Statistical interpretation** Interpretation p-unc = “uncorrected p-value” for Mjob is greater than 0.05. We cannot reject the null hypothesis that each of the mother job results in the same final grade, even after controlling for the student's first term (G1) grade in class. 

**Practical interpretation** We hypothesized that the mother's socioeconomic status (a stable life) would be able to predict a student's final grade. From our results, we see that there is no statistically significant relationship between mother's job and student's final grade when controlling for their first term grade. This raises interesting insights. If we consider mother job category as a proxy for family socioeconomic status, the mother job categories present in the dataset may be too abstract to account for the relationship between a family's socioeconomic status and final grade. The job category of mother's job may not be influencing the student's final grade right now but it is still possible it influences student performance in university (further down the road).
