```python
#ancova example 1
import numpy as np
import pandas as pd

# create data
data = pd.DataFrame({'methodology': np.repeat(['A', 'B', 'C'], 4),
					'current_grade': [67, 88, 75, 85,
									92, 77, 74, 88,
									91, 88, 82, 80],
					'test_score': [77, 89, 74, 69,
									88, 93, 94, 90,
									85, 81, 83, 79]})
# view data
data
```

```python
from pingouin import ancova

data = pd.DataFrame({'methodology': np.repeat(['A', 'B', 'C'], 4),
					'current_grade': [67, 88, 75, 85,
									92, 77, 74, 88,
									91, 88, 82, 80],
					'test_score': [77, 89, 74, 69,
									88, 93, 94, 90,
									85, 81, 83, 79]})

ancova(data=data, dv='test_score', covar='current_grade', between='methodology')
```

```python
#ancova example 2
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Define the data for the ANCOVA
df = pd.DataFrame({'dependent_variable' : [8, 7, 9, 11, 10, 12, 14, 13, 15, 16],
   'group' : ["A", "A", "A", "B", "B", "B", "C", "C", "C", "C"],
    'covariate' : [20, 30, 40, 30, 40, 50, 40, 50, 60, 70]})

# Perform the ANCOVA
model = ols('dependent_variable ~ group + covariate', data=df).fit()

# Print the summary of the model
print(model.summary())
```

```python
#ancova example 3
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/pythonbook/main/Data/clintrial.csv")

pd.crosstab(index=df["drug"], columns=df["therapy"],margins=False)
```

```python
round(df.groupby(['therapy', 'drug'])[['mood_gain']].mean(),2)
```

```python
round(df.groupby(['therapy'])[['mood_gain']].mean(),2)
```

```python
round(df.groupby(['drug'])[['mood_gain']].mean(),2)
```

```python
round(df['mood_gain'].mean(),2)
```

```python
import pingouin as pg
model1 = pg.anova(dv='mood_gain', between='drug', data=df, detailed=True)
round(model1, 2)
```

```python
model2 = pg.anova(dv='mood_gain', between=['drug', 'therapy'], data=df, detailed=True)
round(model2, 4)
```
